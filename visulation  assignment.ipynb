```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate the dataset
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Pivot the data to create a matrix suitable for a heatmap
pivot_table = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')

# Create the heatmap
sns.heatmap(pivot_table, annot=True, fmt=".1f", cmap='YlGnBu')
plt.title('Heatmap of Sales Data')
plt.xlabel('Month')
plt.ylabel('Day')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Flatten the data and create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')


ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show the plot
plt.show()
```

1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
from bokeh.plotting import figure, show, output_notebook
import numpy as np

# Initialize Bokeh notebook output
output_notebook()

# Generate x and y data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a new plot with a title and axis labels
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='y')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="y=sin(x)", line_width=2)

show(p)
```

2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
import numpy as np
import pandas as pd

# Initialize Bokeh notebook output
output_notebook()

# Generate random data
np.random.seed(42)
n = 100
data = {
    'x': np.random.rand(n) * 100,
    'y': np.random.rand(n) * 100,
    'sizes': np.random.rand(n) * 50 + 10,  # Sizes between 10 and 60
    'colors': np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], n)
}
df = pd.DataFrame(data)

# Create a ColumnDataSource from the dataframe
source = ColumnDataSource(df)

# Create a new plot
p = figure(title="Random Scatter Plot", x_axis_label='x', y_axis_label='y')

# Add a scatter renderer with sizes and colors based on the dataframe columns
p.scatter('x', 'y', size='sizes', color='colors', fill_alpha=0.6, source=source)

# Show the results
show(p)
```

3.Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears'] 
counts = [20, 25, 30, 35]

```python



import numpy as np
import pandas as pd

from bokeh.plotting import figure, output_notebook, show
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource
from bokeh.palettes import Spectral6

# Initialize Bokeh notebook output
output_notebook()

# Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
data = {'fruits': fruits, 'counts': counts}
source = ColumnDataSource(data=data)

# Create a new plot
p = figure(x_range=fruits, height=350, title="Fruit Counts",
           toolbar_location=None, tools="")

# Add a bar renderer
p.vbar(x='fruits', top='counts', width=0.9, source=source, legend_field="fruits",
       line_color='white', fill_color=factor_cmap('fruits', palette=Spectral6, factors=fruits))

# Customize plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.y_range.end = 40
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show the plot
show(p)
```

4.Create a Bokeh histogram to visualize the distribution of the given data.


data_hist = np.random.randn(1000) 
hist, edges = np.histogram(data_hist, bins=30)

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()

# Generate the data
data_hist = np.random.randn(1000)

# Compute the histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure
p = figure(title="Histogram", background_fill_color="#fafafa")

# Add a quad glyph to represent the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], 
       fill_color="navy", line_color="white", alpha=0.5)

# Add labels and titl
p.xaxis.axis_label = 'Value'
p.yaxis.axis_label = 'Frequency'
p.grid.grid_line_color="white"
```

```python
show(p)
```

5.Create a Bokeh heatmap using the provided dataset.

data_heatmap = np.random.rand(10, 10) 
x = np.linspace(0, 1, 10) 
y = np.linspace(0, 1, 10) 
xx, yy = np.meshgrid(x, y)


```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import LinearColorMapper, ColorBar
import numpy as np

output_notebook()

# Generate the data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create the Bokeh figure
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y',
           x_range=(0, 1), y_range=(0, 1),
           tooltips=[('Value', '@image')])

# Define the color mapper
color_mapper = LinearColorMapper(palette='Viridis256', low=data_heatmap.min(), high=data_heatmap.max())

# Add the heatmap
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

# Add the color bar
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')
```

```python
show(p)
```
